```python
import re
import random

def repl(m):
    inner_word = list(m.group(2))
    random.shuffle(inner_word)
    return m.group(1) + "".join(inner_word) + m.group(3)

#text = input()
#text = 'По результатам исследований одного английского университета, не имеет значения, в каком порядке расположены буквы в слове. Главное, чтобы первая и последняя буквы были на месте, остальные буквы могут следовать в полном беспорядке, все-равно текст читается без проблем. Причиной этого является то, что мы читаем не каждую букву по отдельности, а все слово целиком.'
text = 'According to the results of the research of one English university, it does not matter in what order the letters in the word are located. The main thing is that the first and last letters are in place, the remaining letters can follow in complete disarray, the text is read without problems. The reason for this is that we do not read each letter individually, but the whole word.'
re.sub(r"(\w)(\w+)(\w)", repl, text)
```
